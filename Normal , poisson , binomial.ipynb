```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm , binom
from scipy.stats import skew
import scipy.stats as stats
import numpy as np
```

# Normal Distribution

Tell weather certain measurements are erroneous or fall within the boundary of expected measures

```python
# Generate dstribution variable
# loc = mean , scale = STD
data_norm = norm.rvs(size = 1000 , loc=50 , scale = 5)
sns.distplot(data_norm , bins = 100 , kde = True)
print(skew(data_norm))
print(np.mean(data_norm))
```

```python
data = stats.poisson.rvs(loc=5,mu=50,size=1500)
sns.distplot(data , bins = 100 , kde = True)
print(skew(data))
print(np.mean(data))
```

```python
# Pdf for normal ditribution
from scipy.stats import norm
from matplotlib import pyplot
pyplot.figure(figsize = (12,4))
mean1 = np.mean(data_norm)
stdev1 = np.std(data_norm)

mean2 = np.mean(data)
stdev2 = np.mean(data)
pdf = norm.pdf(data_norm, mean1, stdev1)
pdf2 = norm.pdf(data, mean2, stdev2)

pyplot.subplot(1,2,1)
pyplot.plot(data_norm, pdf)

pyplot.subplot(1,2,2)
pyplot.plot(data, pdf2)
pyplot.show()
```

# Binomial distribution

Assumptions i) = only 2 possible outcomes

ii) = propability of getting any outcome remains same for all trials

iii) = fixed/finite No.of observations and are independent to each other

```python
# n = no.of obs/trials . p = propability of succes in each trial

data_bin = binom.rvs(n = 10 , p = 0.7 , size = 1000)

sns.displot(data_bin , kde = False)
```
